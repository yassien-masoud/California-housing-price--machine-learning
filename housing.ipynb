```python
#load data
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
test
```

```python
quant=[i for i in data.columns if data.dtypes[i] != 'object']
quant.remove('SalePrice')
quant.remove('Id')
quality=[i for i in data.columns if data.dtypes[i] == 'object']
```

### Numeric Missing Values 

```python
# find the missing values

quant_data_missing=data[quant].isnull().sum()
quant_data_missing=pd.DataFrame(data_quant_missing[data_quant_missing>0])
quant_data_missing
```

```python
# Impute missing values 
from sklearn.impute import SimpleImputer 
my_imputer=SimpleImputer()
impute_numric=pd.DataFrame(my_imputer.fit_transform(data[quant]))
impute_numric.columns=data[quant].columns
impute_numric
```

### Categorical Missing Values

```python
# finding the miss values

qaulity_missing_data= data[quality].isnull().sum() 
qaulity_missing_data= pd.DataFrame(qaulity_missing_data[qaulity_missing_data >0].sort_values(ascending=False))
qaulity_missing_data
```

```python
data[quality].isnull().sum().sum()
```

```python
#filling the massing values
my_impute= SimpleImputer(strategy="most_frequent")
impute_Category= pd.DataFrame(my_impute.fit_transform(data[quality]))
impute_Category.columns= data[quality].columns
impute_Category
```

```python
all_data=pd.concat([impute_Category, impute_numric])
all_data
```

### correlation 

```python
import matplotlib.pyplot as plt
import seaborn as sb
correlation=all_data.corr()
f, ax =plt.subplots(figsize=(30,40))
sb.heatmap(correlation,annot=True)
```
